## Q1.2
Derive the gradient and hessian matrix for the new E(w).

$$\displaystyle E(w) = \frac{1} {2} w ^T \Lambda w - \sum^n_{n=1}[
t_nlny_n +(1-t_n)ln(1-y_n)]$$  
$$ y_n = \sigma(w^TX_n)$$  
$$where \space\sigma(a) = \frac{1} {1+exp(-a)} $$  
$$\frac{\partial {\sigma(a)}}{\partial a} = (-1)(1+exp(-a))^{-2} (-exp(-a))$$  
$$ = \frac{exp(-a)}{[1+exp(-a)]^2} = \sigma(a)\frac{exp(-a)}{1+exp(-a)} $$  
$$ =\sigma(a)(1-\sigma(a)) $$  

$Gradient:$  
$$\displaystyle \nabla E(w) = \frac{\partial {E(w)}}{\partial w} = \Lambda w +\sum^N_{n=1}(y_n-t_n)X_n $$  
$Hessian:$
$$\displaystyle \nabla \nabla E(w) = \Lambda+\sum^N_{n=1}y_n(1-y_n)X_nX_n^T=\Lambda+X^TRX $$  
$$ where \space R_{n,n} = y_n(1-y_n) $$

## Q1.3
Create your mylogistic_l2 class. Show the learned  𝑤  as well as test accuracy for the cases below. If  𝑤  is too long for you, show selected  𝑤  for continuous-valued, binary-valued, and the constant term.  
Case 1: lambda = 1 for all coefficients  
Case 2: lambda = 1 for all but the intercept, no regularization for intercept term.  
Case 3: lambda = 1 for numerical-valued features, lambda = 0.5 for binary-valued features, no regularization for intercept term.  

```python
class mylogistic_l2(object):    
    
    def __init__(self, reg_vec = 1, max_iter = 1000, tol = 10 ** -6, add_intercept = True):
        
        self.numIterations = max_iter
        self.penalty = "L2"
        self.C = reg_vec
        self.tol = 0.1 * tol

    def train2(self, X_train, y_train):

        self.C = np.diag(self.C)
        X_train = np.c_[X_train, np.ones([np.shape(X_train)[0], 1])]
        self.weights = np.dot(np.dot(np.linalg.inv(np.dot(X_train.T, X_train) + np.mean(self.C) * np.eye(X_train.shape[1])),X_train.T),y_train)
        
        
        for i in range(self.numIterations):
            
            z = np.dot(X_train, self.weights)
            y_result = logistic_func(z)
            errors = y_result - y_train
            
            delta_w = np.dot(X_train.T, errors) + np.dot(self.C, self.weights)
            
            rnn = y_result * (1 - y_result)
            hessian = self.C + np.dot((X_train.T * rnn), X_train)
                
            self.iterationsPerformed = i
            
            delta = np.dot(np.linalg.inv(hessian), delta_w)

            if np.all(abs(delta) >= self.tol): 
                #weight update
                self.weights -= delta
            else:
                break
            
        return self
                    
    def predict(self, X_test, pi = 0.5):
       
        z = np.dot(X_test, self.weights[0:-1]) + self.weights[-1]
        probs = np.array([logistic_func(i) for i in z])
        predictions = np.where(probs >= pi, 1, 0)
       
        return predictions, probs

def logistic_func(z):   
    return 1 / (1 + np.exp(-z)) 
    
```
